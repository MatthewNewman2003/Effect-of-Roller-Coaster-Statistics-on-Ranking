**Importing Libraries**

```python
#Importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from scipy.stats import spearmanr
```

**Reading in Data**

```python
#Declaring Google Drive URL
GoogleDriveURL = "https://docs.google.com/spreadsheets/d/1OfxlATMBJJwN1VjW7sb0jKwIFywJcACbZYJHIxbjdVc/export?format=csv&gid=0"

#Reading the CSV file into a Pandas DataFrame
Data = pd.read_csv(GoogleDriveURL)
```

```python
#Printing the head of the dataset
print(Data.head())
```

**Relationship between Height and Ranking**

```python
#Assigning the Ranking and Height columns to variables
Ranking = Data["Ranking"]
Height = Data["Height"]

#Creating a scatter graph to show the relationship between Height and Ranking
plt.scatter(Height, Ranking)
plt.title("Relationship between Height and Ranking")
plt.xlabel("Height (ft)")
plt.ylabel("Ranking")
plt.show()
```

```python
#Creating an array to pass to the correlation tests
CombinedArray = np.vstack([Ranking, Height])
CombinedArray = CombinedArray[:,~np.any(np.isnan(CombinedArray), axis=0)]
```

Pearson's Correlation Coefficient

```python
#Performing a Pearson's correlation hypothesis test
Correlation, PValue = pearsonr(CombinedArray[0], CombinedArray[1])

#Denoting the strength of the correlation based on the absolute value of its coefficient
if abs(Correlation) < 0.3:
    print("The Pearson correlation coefficient between Ranking and Height is {}, indicating no significant Pearson correlation.".format(Correlation))
elif 0.3 < abs(Correlation) < 0.5:
    print("The Pearson correlation coefficient between Ranking and Height is {}, indicating a weak Pearson correlation.".format(Correlation))
elif 0.5 < abs(Correlation) < 0.7:
    print("The Pearson correlation coefficient between Ranking and Height is {}, indicating a moderate Pearson correlation.".format(Correlation))
elif 0.7 < abs(Correlation) < 0.9:
    print("The Pearson correlation coefficient between Ranking and Height is {}, indicating a strong Pearson correlation.".format(Correlation))
else:
    print("The Pearson correlation coefficient between Ranking and Height is {}, indicating a very strong Pearson correlation.".format(Correlation))

#Denoting the degree of evidence for a non-zero correlation based on the p-value
if PValue < 0.01:
    print("The p-value is: {}, indicating extremely significant evidence of a non-zero correlation under Pearson's r.".format(PValue))
elif 0.01 < PValue < 0.05:
    print("The p-value is: {}, indicating significant evidence of a non-zero correlation under Pearson's r.".format(PValue))
elif 0.05 < PValue < 0.1:
    print("The p-value is: {}, indicating marginally significant evidence of a non-zero correlation under Pearson's r.".format(PValue))
else:
    print("The p-value is: {}, indicating insufficient evidence of a non-zero correlation under Pearson's r.".format(PValue))
```

Spearman's Correlation Coefficient

```python
#Performing a Spearman's correlation hypothesis test
Correlation, PValue = spearmanr(CombinedArray[0], CombinedArray[1])

#Denoting the strength of the correlation based on the absolute value of its coefficient
if abs(Correlation) < 0.3:
    print("The Spearman's Rank correlation coefficient between Ranking and Height is {}, indicating no significant Spearman correlation.".format(Correlation))
elif 0.3 < abs(Correlation) < 0.5:
    print("The Spearman's Rank correlation coefficient between Ranking and Height is {}, indicating a weak Spearman correlation.".format(Correlation))
elif 0.5 < abs(Correlation) < 0.7:
    print("The Spearman's Rank correlation coefficient between Ranking and Height is {}, indicating a moderate Spearman correlation.".format(Correlation))
elif 0.7 < abs(Correlation) < 0.9:
    print("The Spearman's Rank correlation coefficient between Ranking and Height is {}, indicating a strong Spearman correlation.".format(Correlation))
else:
    print("The Spearman's Rank correlation coefficient between Ranking and Height is {}, indicating a very strong Spearman correlation.".format(Correlation))

#Denoting the degree of evidence for a non-zero correlation based on the p-value
if PValue < 0.01:
    print("The p-value is: {}, indicating extremely significant evidence of a non-zero correlation under Spearman's Rank.".format(PValue))
elif 0.01 < PValue < 0.05:
    print("The p-value is: {}, indicating significant evidence of a non-zero correlation under Spearman's Rank.".format(PValue))
elif 0.05 < PValue < 0.1:
    print("The p-value is: {}, indicating marginally significant evidence of a non-zero correlation under Spearman's Rank.".format(PValue))
else:
    print("The p-value is: {}, indicating insufficient evidence of a non-zero correlation under Spearman's Rank.".format(PValue))
```
